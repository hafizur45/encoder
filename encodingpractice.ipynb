Using loop

```python
df4.head()
```

```python
for col in df4.columns:
  if df4[col].dtypes == 'object':
    dum = pd.get_dummies(df4[col],prefix=col)
    df4.drop(col,axis=1,inplace=True)
    df4=pd.concat([df4,dum],axis=1)
```
